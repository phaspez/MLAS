````python
from zlib import crc32
def is_id_in_test_set(id, test_ratio):
    return crc32(np.int64(id)) < test_ratio * 2**32

def split_data_with_id_hash(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: is_id_in_test_set(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
````

##### scikit-learn design



1. Consistency

All objects share a consistent and simple interface:

**Estimators** are objects that can estimate some parameters on a dataset. `SimpleImputer `is an estimator. Estimations are performed by the `fit()` method , it takes a dataset as the parameter, two if supervised (second is the labels), any other parameters to guide the estimator are **hyperparameters**, and they must usually be set via a constructor parameter.

**Transformers** are estimators that can transform a dataset. Transforms are performed by the `transform()` method, with the dataset to transform as the parameter, it returns the transformed dataset. It also have a `fit_transform() `method, which is equivalent to call `fit() `and then `transform(),` sometimes it's more optimized and runs faster.

**Predictors **are estimators that are capable of making predictions, e.g. `LinearRegression` model. Have a `predict()` method that takes a dataset of new instances and return a dataset of corresponding dataset of predictions. Have a `score()` method that measures the quality of predictions, given the test set (and corresponding labels, if supervised learning).

2. Inspection

Estimator's hyperparameters are accessible directly via public instances, e.g. `imputer.strategy, `and learned parameters are accessible via public instances with underscore suffix, e.g. `imputer.statistics_.`

3. Nonproliferation of classes

Datasets are represented as NumPy or SciPy sparse matrices. Hyperparamters are regular Python strings or numbers.

4. Composition

It's easy to create a `Pipeline` estimatorfrom an arbitrary sequence of transformers followed by a final estimator.

5. Sensible defaults

scikit-learn provides resonable default values for most parameters.

````python
# custom transformer that acts like the StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_array, check_is_fitted

class StandardScalerClone(BaseEstimator, TransformerMixin):
    def __init__(self, with_mean=True):
        self.with_mean = with_mean
    
    def fit(self, X, y=None):
        X = check_array(X)
        self.mean_ = X.mean(axis=0)
        self.scale_ = X.std(axis=0)
        self.n_features_in_ = X.shape[1]
        return self
    
    def transform(self, X):
        check_is_fitted(self)
        X = check_array(X)
        assert self.n_features_in == X.shape[1]
        if self.with_mean:
            X = X - self.mean_
        return X / self.scale_
````

````python
# custom transformer that use another estimator in its impl 
from sklearn.cluster import KMeans
class ClusterSimilarity(BaseEstimator, TransformerMixin):
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state
    
    def fit(self, X, y=None, sample_weight=None):
        self.kmeans_ = KMeans(self.n_clusters, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self
    
    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)
    
    def get_feature_names_out(self, names=None):
        return [f"Cluster {i} similarity" for i in range(self.n_clusters)]
    
````

````python
from sklearn.compose import ColumnTransformer
num_attribs = ["longitude", "latituude", "housing_median_age", "total_rooms", "total_bedrooms", "population", "households", "median_income"]
cat_attribs = ["ocean_proximity"]
cat_pipeline = make_pipeline(SimpleImputer(strategy="most_frequent"), OneHotEncoder(handle_unknown="ignore"))

preprocessing = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs),
    # you can specify 'drop' if you want the columns to be dropped
    # ("example", "drop", attribs)
    # by default, any remaining columns not listed will be dropped
    # specify 'passthrough' to left it untouched
])
````

````python
# another way of setting column transformer. 
from sklearn.compose import make_column_selector, make_column_transformer
import numpy as np

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)), # this will select all numeric attr
    (cat_pipeline, make_column_selector(dtype_include=object)), # this select all string
)
````

````python
# divide the first column by the second column
def column_ratio(X):
    return X[:, [0]] / X[:, [1]]

def ratio_name(function_transformer, feature_names_in):
    return ["ratio"]

def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler()
    )

log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler()
)

cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)

default_num_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler(),
)

preprocessing = ColumnTransformer(
    transformers=[
        ("rooms_per_house", ratio_pipeline(), ["population", "households"]),
        ("log", log_pipeline, ["total_bedrooms", "total_rooms", "population", "households", "median_income"]),
        ("geo", cluster_simil, ["latitude", "longitude"]),
        ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
    ],
    remainder=default_num_pipeline
)
````

### Fine tuning model



#### GridSearchCV



````python
from sklearn.model_selection import GridSearchCV

full_pipeline = Pipeline([
    ("preprocessing", preprocessing),
    ("random_forest", RandomForestRegressor(random_state=42)),
])

param_grid = [
    {
        'preprocessing__geo__n_clusters': [5, 8, 10],
        'random_forest__max_features': [4, 6, 8]
    }, {
        'preprocessing__geo__n_clusters': [10, 15],
        'random_forest__max_features': [6, 8, 10]
    },
]

grid_search = GridSearchCV(full_pipeline, param_grid, cv=3, scoring="neg_root_mean_squared_error")

grid_search.fit(housing, housing_labels)
````
